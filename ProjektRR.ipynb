## 2. Modelowanie epidemii

<div style="text-aligin: justify">
Matematyczne modelowanie epidemii pozwala na analizę dynamiki rozprzestrzeniania się chorób zakaźnych. Modele te opierają się na podziale populacji na grupy, co umożliwia analizę wzajemnych przejść między tymi grupami w czasie. Jednym z podstawowych podejść do problemu jest wykorzystanie modeli przedziałowych, takich jak model SIR i jego modyfikacje, które pozwalają prognozować rozwój epidemii oraz ocenić skuteczność interwencji.
</div>

### 2.1 Klasyczny model SIR

<div style="text-aligin:justify">
Model SIR, zaproponowany przez Kermacka i McKendricka w 1927 roku, dzieli populacje na trzy grupy:

- S (Susceptible) - osoby zdrowe, podatne na zakażenia,
- I (Infected) - osoby zakażone, mogące zarazić innych,
- R (Recovered/Removed) - osoby, które wyzdrowiały, zostały odizolowane lub zmarły.

Dynamikę epidemii w tym modelu opisuje układ równań różniczkowych
\begin{equation*}
    \left\{\begin{split}
        &\frac{dS}{dt}=-\beta SI \\
        &\frac{dI}{dt}=\beta SI -\alpha I\\
        &\frac{dR}{dt}=\alpha I,
    \end{split}\right.
\end{equation*}
gdzie $\beta$ to współczynnik zakażeń, określających tempo transmisji wirusa, a $\alpha$ jest współczynnikiem ozdrowień, odpowiadający za tempo przejścia osób z grupy $I$ do grupy $R$.

Przyjmuje się, że populacja jest stała (tj. $N=S+I+R$), co oznacza brak migracji, narodzin oraz zgonów z przyczyn innych niż choroba.
</div>

#### 2.2 Współczynnik ozdrowienia $R_0$

<div style="text-aligin: justify">

Podstawowym parametrem charakteryzującym epidemię jest współczynnik odnowienia $R_0$, definiowany jako średnia liczba osób, które może zarazić jedna osoba w pełni podanej populacji. Jeśli $R_0>1$ to liczba zakażeń rośnie co prowadzi do epidemii. W przypadku, gdy $R_0< 1$ epidemia wygasa. 

</div>

### 2.3 Model SEIR

<div style="text-aligin:justify">

Dla chorób, które charakteryzują się okresem inkubacji, klasyczny model SIR jest rozwijany do modelu SEIR. Dodaje on grupę

- E (Exposed): osoby zarażone, ale jeszcze niezarażające innych.

Układ równań modelu SEIR uwzględnia okres latencji zapisujemy jako

\begin{equation*}
    \left\{\begin{split}
        &\frac{dS}{dt}=-\beta SI \\
        &\frac{dE}{dt}=-\beta SI - \gamma E \\
        &\frac{dI}{dt}=\gamma - \alpha I \\
        &\frac{dR}{dt}=\alpha I,
    \end{split}\right.
\end{equation*}
gdzie $\gamma$ to współczynnik przejścia z grupy E do grupy I. 

Model SEIR lepiej odzwierciedla rzeczywiste procesy epidemiologiczne, umożliwiając uwzględnienie okresów bezobjawowego zakażenia

</div>

### 2.4 Zastosowanie modeli

<div style="text-aligin:justify">

Modele takie jak SIR i SEIR znalazły szerokie zastosowanie podczas epidemii COVID-19. Używano ich do prognozowania liczby przypadków, oceniania skuteczności działań prewencyjnych, takich jak dystans społeczny oraz planowanie zasobów służby zdrowia. Dynamika epidemii przedstawiona za pomocą tych modeli dostarcza istotnych informacji, takich jak moment osiągnięcia szczytu zachorowań czy potencjalne skutki zmiany parametrów modelu, co czyni je nieocenionym narzędziem w zarządzaniu epidemii <a href="#bibliografia4"><cite>[4]</cite></a>.

</div>

## 3. Zastosowanie modelu SEIR w modelowaniu epidemii palenia papierosów w Polsce

### 3.1 Oznaczenia

<div style="text-aligin:justify">

Epidemiologia nałogów, takich jak palenie papierosów, wykazuje wiele podobieństw do klasycznych epidemii. Palenie rozprzestrzenia się w sposób społeczny, gdzie kontakt z osobami palącymi może wywierać wpływ na osoby podatne, które zaczynają eksperymentować i w końcu przechodzą do aktywnego nałogu.

W naszym modelu dzielimy populację $P(t)$ na cztery różne grupy:

- $N(t)$ - osoby niepalące, ale podatne,
- $O(t)$ - osoby palące okazjonalnie (palące mniej niż paczkę papierosów dziennie),
- $M(t)$ - osoby palące nałogowo (palące więcej niż paczkę papierosów dziennie),
- $B(t)$ - osoby, które rzuciły palenie, jednak mogące wrócić do nałogu.

Populacja wyraża się wzorem $P(t)=N(t)+O(t)+M(t)+B(t)$.

Współczynniki śmiertelności są oznaczane:
- $d_N$ - osób niepalących,
- $d_O$ - osób palących okazjonalnie,
- $d_M$ - osób palących nałogowo,
- $d_B$ - osób, które rzuciły palenie.

1\. Zmiany w grupie osób niepalących:
- Niepalący dołączają do populacji z uwagi na stały wskaźnik urodzeń - $\mu P(t)$,
- Liczba niepalących maleje z powodu śmierci naturalnej $d_N N(t)$,
- Niepalący mogą stać się palaczami w wyniku interakcji z innymi palaczami z tempem $\beta \frac{O(t)+M(t)} {P(t)} $.  

2\. Zmiany w grupie osób palących okazjonalnie:
- Palacze okazjonalni przybywają, gdy niepalący zaczynają palić z tempem $\beta \frac{O(t)+M(t)} {P(t)} $,
- Nałogowi palacza mogą zmniejszyć ilość spożywanych papierosów i dołączyć do grupy okazjonalnych palaczy z szybkością $\alpha M(t)$,
- Okazjonalni palacze mogą stać się nałogowymi palaczami z intensywnością $\gamma O(t)$,
- Osoby mogą również rzucić palenie przechodząc do grupy byłych palaczy z tempem $\lambda O(t)$,
- Byli palacze mogą powrócić do palenia w tempie $\rho B(t)\frac{O(t)+M(t)} {P(t)} $.  

3\. Zmiany w grupie osób palących nałogowo:
- Palacze okazjonalni mogą palić więcej i przejść do grupy nałogowych palaczy z szybkością $\gamma O(t)$,
- Ubytek nałogowych palaczy następuje, gdy zmniejszą ilość spożywanych papierosów z tempem $\alpha M(t)$,
- Nałogowi palacze mogą rzucić palenie z intensywnością $\delta M(t)$.  

4\. Zmiany w grupie osób byłych palaczy:
- Osoby, które rzuciły palenia są byłymi okazjonalnymi i nałogowymi palaczami, którzy dołączają do tej grupy z intensywnością $\delta M(t)+\lambda O(t)$,
- Spadek jest wynikiem powrotu do palenia w tempie $\rho B(t)\frac{O(t)+M(t)} {P(t)} $. 


</div>

### 3.1 Założenia modelu

<div style="text-aligin:justify">

1\. Parametry przejść między grupami są stałe w czasie, ponieważ analizujemy model w krótkim okresie.  
2\. Zmiany liczebności subpopulacji wpływają dynamicznie na zachowanie związane z paleniem.  
3\. Przejście między grupami są proporcjonalne do ich liczebności.  
4\. Przejście z grupy niepalących do klasy palaczy jest spowodowane wpływem społecznym palaczy.  
5\. Śmiertelność niezależna od populacji różni się dla każdej grupy: $d_N<d_B<d_O<d_M$.  
6\. Osoby, które rzuciły palenie mogą powrócić do nałogu.  
7\. Osoba jest uważana za nałogowego palacza, jeżeli pali więcej niż paczkę papierosów dziennie.  

</div>

### 3.2 Równania modelu


Bazując na założeniach można ułożyć następujący układ równań różniczkowych
\begin{equation*}
    \left\{\begin{split}
        &\frac{dN}{dt}=\mu P(t)-\beta N(t)\frac{O(t)+M(t)}{P(t)}-d_N N(t) \\
        &\frac{dO}{dt}=\beta N(t)\frac{O(t)+M(t)}{P(t)}+\alpha M(t)+ \rho B(t)\frac{O(t)+M(t)}{P(t)}-\gamma O(t)-\lambda M(t)-d_O O(t) \\
        &\frac{dM}{dt}=\gamma O(t)-\alpha M(t)-\delta M(t)-d_M M(t) \\
        &\frac{dB}{dt}=\lambda O(t)+\delta M(t)-\rho B(t)\frac{O(t)+M(t)}{P(t)}-d_B B(t).
    \end{split}\right.
\end{equation*}
Na podstawie powyższego układu równań możemy uzyskać schemat <i><b>(Schemat 1.)</b></i>, który odzwierciedla zmiany w poszczególnych grupach. Kierunek przepływu osób między grupami są reprezentowane przez skierowane strzałki.

W modelu SEIR współczynnik $R_0$ odnosi się do średniej liczby przypadków, jakie generuje osobnik w pełni podatny na danym etapie epidemii, w populacji całkowicie podatnej. Jest to kluczowy parametr, który opisuje dynamikę rozprzestrzeniania się epidemii. W kontekście epidemii palenia papierosów w Polsce $R_0$ można zinterpretować jako wskaźnik, który określa, ile nowych osób zaczyna palić papierosy w wyniku kontaktu z jednym aktywnym palaczem w populacji. ## 4. Metody numeryczne

W tej sekcji zajmiemy się wprowadzeniem i zdefiniowaniem metody Eulera oraz Runge-Kutty i wybraniem dokładniejszego sposobu w celu dalszej analizy wyników.

### 4.1 Metoda Eulera

Do wprowadzenia metody numerycznej Eulera potrzebujemy najpierw skorzystać z podstawowego twierdzenia <a href="#bibliografia5"><cite>[5]</cite></a> dotyczącego istnienia i jednoznaczności rozwiązania równań różniczkowych

#### Twierdzenie 1 
Niech funkcja $f$ i $\frac{\text{d}y} {\text{d}t} $ będą ciągłe na prostokącie $\alpha<t<\beta$, $\gamma<y<\delta$ zawierający punkt $(t_ {0}, y_ {0}) $. Na pewnym przedziale $t_ {0} -h<t<t_ {0} +h$ zawartym w $\alpha<t<\beta$ istnieje jednoznaczne rozwiązanie $y=\phi (t)$ z warunkami początkowymi  
$$y'=f (t, y), \, \, \, \, y(t_ {0})=y_0.$$

Twierdzenie to zapewnia podstawę teoretyczną dla stosowania metod numerycznych, takich jak metoda Eulera, umożliwiając przybliżenie rozwiązania problemu początkowego
$$\frac{\text{d}y}{\text{d}t} = f(t,y),\,\,\,\,y(t_{0})=y_{0}.$$

Metoda numeryczna Eulera <a href="#bibliografia5"><cite>[5]</cite></a> jest jedną z najprostszych metod przybliżeniowego rozwiązywania równań różniczkowych. Opiera się na przybliżaniu pochodnej za pomocą różnic skończonych i przybliżeniu wartości funkcji w kolejnych krokach za pomocą wzoru iteracyjnego

$$y_{n+1}=y_n + f(t_n,y_n)(t_{n+1}-t_n),\,\,\, n =0,1,2,...$$  

Jeśli krok czasowy ma stałą wartość $h$ i oznaczymy $f(t_n,y_n)$ za pomocą $f_n$ powyższe równanie możemy uprościć do  
$$y_{n+1}=y_n + f_nh,\,\,\, n =0,1,2,...$$  

Metoda Eulera zakłada wielokrotne obliczanie powyższego równania wykorzystując poprzedni krok do wyliczenia następnego. Dzięki tej procedurze uzyskujemy wartości $y_1,y_2,...,y_n,...$, które przybliżają wartość rozwiązania $\phi(t)$ w punktach $t_1,t_2,...,t_n,...$ Metodę Eulera można łatwo zaimplementować w różnych językach programowania. W naszym projekcie skupimy się na implementacji w języku Python. Wykorzystamy następujące kroki:  
<b>Krok 1.</b> zdefiniowanie $f(t,y)$,  
<b>Krok 2.</b> podanie warunków początkowych $t_0$ i $y_0$,  
<b>Krok 3.</b> podanie rozmiaru kroku $h$ i liczby kroków $n$,  
<b>Krok 4.</b> wypisanie $t_0$ i $y_0$,  
<b>Krok 5.</b> <b>dla</b> $j$ z przedziału od $1$ do $n$ <b>wykonaj</b>  
$k_1=f(t,y)$  
$y=y+h\cdot k_1$  
$t=t+h$  
<b>Krok 6.</b> wypisanie $t$ i $y$,  
<b>Krok 7.</b> zakończ.  
Implementacja w języku python wygląda następująco

```python
def metoda_eulera():
    # Krok 1: Definicja funkcji f(t, y)
    def f(t, y):
        return y #dowolna funkcja f(t,y)

    # Krok 2: Warunki początkowe
    t0 = t0
    y0 = y0 

    # Krok 3: Rozmiar kroku h i liczba kroków n
    h = h
    n = n
    # Krok 4: Wypisanie t0 i y0
    print(f"t0={t0}, y0={y0}")

    # Zmienne pomocnicze
    t = t0
    y = y0

    # Krok 5: Iteracja metody Eulera
    for j in range(1, n + 1):
        k1 = f(t, y)
        y += h*k1
        t += h

        # Krok 6: Wypisanie wyników dla aktualnego kroku
        print(f"Krok {j}: t={t}, y={y}")
```

### 4.2 Metoda Runge-Kutty

Metoda Runge-Kutty jest jedną z najpopularniejszych metod numerycznych stosowanych do rozwiązywania równań różniczkowych zwyczajnych, szczególnie gdy nie da się znaleźć rozwiązania analitycznego. Również korzystamy tutaj z <i>Twierdzenia 1</i> o jednoznaczności rozwiązania równania różniczkowego z warunkami początkowymi  
$$\frac{\text{d}y}{\text{d}t} = f(t,y),\,\,\,\,y(t_{0})=y_{0}.$$
Jest to metoda <a href="#bibliografia5"><cite>[5]</cite></a> wielokrokowa, w której iteracyjnie przybliża się rozwiązanie poprzez uwzględnienie wartości pochodnych w kilku punktach w obrębie jednego kroku. Najczęściej stosowaną odmianą jest metoda czwartego rzędu, która oferuje korzystny kompromis między dokładnością a obciążeniem obliczeniowym. Ogólny schemat metody można zapisać jako  
\begin{equation*}
y_{n+1}=y_n+h\left(\frac{k_{n1}+2k_{n2}+2k_{n3}+k_{n4}}{6}\right),
\end{equation*}  
gdzie
\begin{equation*}
    \begin{split}
        &k_{n1}=f(t_n,y_n),\\
        &k_{n2}=f(t_n+\tfrac{1}{2}h,y_n+\tfrac{1}{2}hk_{n1}),\\
        &k_{n3}=f(t_n+\tfrac{1}{2}h,y_n+\tfrac{1}{2}hk_{n2}),\\
        &k_{n4}=f(t_n+h,y_n+hk_{n3}).
    \end{split}
\end{equation*}  
Suma $\frac{1}{6}(k_{n1}+2k_{n2}+2k_{n3}+k_{n4})$ może być interpretowana jako średnie nachylenie. Implementacja tej metody nie jest mocno skomplikowana, wymaga tylko zmiany kroku 5 z poprzedniego schematu z podsekcji <i>4.1 Metoda Eulera</i> na  
<b>Krok 5.</b>  
$k1=f(t,y)$,  
$k2=f(t+0.5h,y+0.5hk1)$,  
$k3=f(t+0.5h,y+0.5hk2)$,  
$k2=f(t+h,y+k3)$,  
$y=y+\tfrac{h}{6}\cdot(k1+2k2+2k3+k4)$,  
$t=t+h$.  
Metoda Runge-Kutty tak samo jak Eulera ma problem z bardzo małymi krokami czasowymi $h$. Jednak błąd jest wystarczająco mały, aby uzyskać satysfakcjonujące nas przybliżenie funkcji analitycznej. Tę metodę implementujemy do języka Python w następujący sposób

```python
def metoda_runge_kutty():
    # Krok 1: Definicja funkcji f(t, y)
    def f(t, y):
        return y  # dowolna funkcja f(t, y)

    # Krok 2: Warunki początkowe
    t0 = t0
    y0 = y0

    # Krok 3: Rozmiar kroku h i liczba kroków n
    h = h
    n = n

    # Krok 4: Wypisanie t0 i y0
    print(f"t0={t0}, y0={y0}")

    # Zmienne pomocnicze
    t = t0
    y = y0

    # Krok 5: Iteracja metody Rungego-Kutty
    for j in range(1, n+1):
        k1 = f(t, y) 
        k2 = f(t+0.5*h, y+0.5*h*k1)
        k3 = f(t+0.5*h, y+0.5*h*k2)
        k4 = f(t+h, y+h*k3)
        y += (h/6)*(k1+2*k2+2*k3+k4)
        t += h 

        # Krok 6: Wypisanie wyników dla aktualnego kroku
        print(f"Krok {j}: t={t}, y={y}")
```

### 4.3 Porównanie metod

W tej sekcji dokonamy porównania metody Eulera z metodą Runge-Kutty, w tym celu porównamy zbieżność, rząd metod oraz ich błędy. Wprowadźmy najpierw dwie definicje <a href="#bibliografia5"><cite>[5]</cite></a>.  
#### Definicja 1 (lokalny błąd obcięcia)  
Rozważmy równanie różniczkowe z warunkiem początkowym postaci 
$$y'=f(t,y),\,\,y(t_n)=y_n.$$
Niech $y(t_{n+1})$ oznacza dokładne rozwiązanie w punkcie $t_{n+1}$, a $y_{n+1}$ oznacza rozwiązanie numeryczne dla danej metody. Wtedy lokalnym błędem obcięcia (LTE, <i>Local Truncation Error</i>) nazywamy
$$\text{LTE}=y(t_{n+1})-y_{n+1}.$$ 
#### Definicja 2 (globalny błąd obcięcia)
Niech $y(t_n)$ oznacza dokładne rozwiązanie równania różniczkowego w punkcie $t_n$, a $y_n$ wynik z metody numerycznej, wtedy globalnym błędem obcięcia (GTE, <i>Global Truncation Error</i>) nazywamy  
$$\text{GTE}=|y(t_n)-y_n|.$$ 
Następnie podamy kilka faktów, które umożliwią nam dokładniejszą analizę
#### Fakt 1 (rząd metody)  
Globalny błąd obcięcia GTE, który mierzy błąd na całym przedziale czasowym, jest wynikiem sumowania błędów lokalnych w kolejnych krokach. Dla metody o rzędzie $p$, globalny błąd obcięcia jest proporcjonalny do $O(h^p)$, gdzie $h$ to krok czasowy, więc  
$$\text{GTE}\approx O(h^p).$$  
Za to LTE, ponieważ jest błędem popełnionym w jednym kroku, dla metody numerycznej o rzędzie $p$, LTE jest proporcjonalny do $O(h^{p+1})$.

Za pomocą <i>Definicji 1 </i> zbadajmy LTE dla metody Eulera  
$$y_{n+1}=y_n + hf(t_n,y_n),$$
gdzie $h$ jest rozmiarem kroku czasowego. Podstawmy dokładne rozwiązanie $y(t_n)$ za $y_n$ i rozwiniemy $y_{n+1}$ w szereg Taylora wokół $t_n$  
$$y(t_{n+1})=y(t_n)+hy'(t_n)+\frac{1}{2}h^2y''(\zeta),$$
gdzie $\zeta$ jest z przedziału ($t_n,t_{n+1}$). Podstawiamy tę wartość do definicji i skorzystajmy z faktu, że $y'(t_n)=f(t_n,y(t_n))$  
$$\text{LTE}=y(t_n)+hy'(t_n)+h^2y''(\zeta) - y_{n+1}=y_n + hf(t_n,y_n) = h^2y''(\zeta)$$  
Stąd mamy, że lokalny błąd obcięcia jest proporcjonalny do kwadratu kroku $h$, tj. $\text{LTE} \sim O(h^2)$. 

Powtórzmy całą procedurę dla metody numeryczne Runge-Kutty zdefiniowanej jako 
\begin{equation*}
y_{n+1}=y_n+h\left(\frac{k_{n1}+2k_{n2}+2k_{n3}+k_{n4}}{6}\right),
\end{equation*}  
gdzie
\begin{equation*}
    \begin{split}
        &k_{n1}=f(t_n,y_n),\\
        &k_{n2}=f(t_n+\tfrac{1}{2}h,y_n+\tfrac{1}{2}hk_{n1}),\\
        &k_{n3}=f(t_n+\tfrac{1}{2}h,y_n+\tfrac{1}{2}hk_{n2}),\\
        &k_{n4}=f(t_n+h,y_n+hk_{n3}).
    \end{split}
\end{equation*}  
Podstawiamy tak samo dokładne rozwiązanie za rozwiązanie numeryczne i rozwijamy w szereg Taylora wokół $t_n$ 
$$y(t_{n+1})=y(t_n)+hy'(t_n)+\frac{1}{2}h^2y''(t_n)+\frac{1}{6}h^3y^{(3)}(t_n)+\frac{1}{24}h^4y^{(4)}(t_n)+\frac{1}{120}h^5y^{(5)}(\zeta).$$  
Należy teraz zauważyć, że z definicji tej metody musimy również rozwinąć w szereg Taylora wyrazy $k_{ni}$ dla $i\in \{1,2,3,4\}.$ Wykonując tę czynność oraz podstawiając gotowe wyniki do definicji błędu otrzymujemy, że 
$$\text{LTE}=\frac{1}{120}h^5y^{(5)}(\zeta).$$
Stąd mamy, że metoda Runge-Kutty jest obarczona błędem rzędu $h^5$.

```python
import numpy as np
import matplotlib.pyplot as plt

def dN_dt(N, O, M, B):
    return mu*P(N, O, M, B)-beta*N*((O+M)/P(N, O, M, B))-d_n*N

def dO_dt(N, O, M, B):
    return (beta*N*((O+M)/P(N, O, M, B))+alfa*M+rho*B*((O+M)/P(N, O, M, B))-gamma*O-lambd*M-d_o*O)

def dM_dt(N, O, M, B):
    return gamma*O-alfa*M-delta*M-d_m*M

def dB_dt(N, O, M, B):
    return lambd*O+delta*M-rho*B*((O+M)/P(N, O, M, B))-d_b*B

def P(N, O, M, B):
    return N+O+M+B

mu = 0.0101
d_n = 0.0108
d_o = 0.0216
d_m = 0.0324
d_b = 0.0140
beta = 0.443
alfa = 0.105
gamma = 0.243
lambd = 0.365
rho = 0.1215
delta = 0.186

def rk4(f, t, y, h):
    k1 = f(t, y) 
    k2 = f(t+0.5*h, y+0.5*h*k1)
    k3 = f(t+0.5*h, y+0.5*h*k2)
    k4 = f(t+h, y+h*k3)
    return y+(h/6)*(k1+2*k2+2*k3+k4)

def rozwiazanie_ukladu(t0, t_max, h, N0, O0, M0, B0):
    wartosc_t = []
    wartosc_N = []
    wartosc_O = []
    wartosc_M = []
    wartosc_B = []

    t = t0
    N, O, M, B = N0, O0, M0, B0

    while t <= t_max:
        wartosc_t.append(t)
        wartosc_N.append(N)
        wartosc_O.append(O)
        wartosc_M.append(M)
        wartosc_B.append(B)

        N = rk4(lambda t, N: dN_dt(N, O, M, B), t, N, h)
        O = rk4(lambda t, O: dO_dt(N, O, M, B), t, O, h)
        M = rk4(lambda t, M: dM_dt(N, O, M, B), t, M, h)
        B = rk4(lambda t, B: dB_dt(N, O, M, B), t, B, h)

        t += h

    return wartosc_t, wartosc_N,  wartosc_O, wartosc_M, wartosc_B

t0 = 0
t_max = 200
h = 0.1
N0 = 25735469
O0 = 4106919
M0 = 4351982
B0 = 4225226

wartosc_t, wartosc_N, wartosc_O, wartosc_M, wartosc_B = rozwiazanie_ukladu(t0, t_max, h, N0, O0, M0, B0)

plt.figure(figsize=(12, 8))

plt.plot(wartosc_t, wartosc_N, label='N(t)', color='blue')
plt.plot(wartosc_t,  wartosc_O, label='O(t)', color='green')
plt.plot(wartosc_t, wartosc_M, label='M(t)', color='red')
plt.plot(wartosc_t, wartosc_B, label='B(t)', color='orange')

plt.xlabel('Czas t')
plt.ylabel('Wartości (skala logarytmiczna)')
plt.yscale('log')
plt.title('Wykres 4. rozwiązanie układu równań różniczkowych')
plt.legend()
plt.grid()
plt.show()
```

## 6. Wnioski

<div style="text-aligin:justify">

1\. <b>Zmniejszający się odsetek codziennych palaczy. </b> W latach 2009-2019 zauważono spadek liczby osób deklarujących codzienne palenie papierosów z 31% do 21%. Ten trend spadkowy świadczy o skuteczności wprowadzonych polityk zdrowotnych, takich jak kampanie antynikotynowe, wzrost akcyzy na wyroby tytoniowe czy zakaz palenia w miejscach publicznych. Pomimo tego problem palenia wciąż pozostaje istotny, szczególnie w kontekście grup o wyższej skłonności od uzależnienia.  
2\. <b>Różnice między płciami. </b> Analiza ujawniła, że w 2019 roku większy odsetek mężczyzn (24%) niż kobiet (18%) deklarowało codzienne palenie. Różnica ta może wynikać z uwarunkowań kulturowych oraz różnic w stylu życia między płciami. Niemniej jednak wzrost liczby palących kobiet w niektórych grupach wiekowych wymaga szczególnej uwagi.  
3\. <b>Grupy wiekowe najbardziej narażone na uzależnienie. </b> Dane wskazują, że regularne palenie wzrasta z wiekiem, osiągając szczyt w grupie wiekowej 30-40 lat. Szczególnie niepokojący jest wzrost odsetka palących kobiet w tej grupie, co może wiązać się z czynnikami środowiskowymi, stresem zawodowym lub społecznymi oczekiwaniami.  
4\. <b>Niska liczba palącej młodzieży - pozytywny sygnał. </b> Wśród młodzieży w wieku 15-19 lat codziennie palenie deklaruje jedynie 3% dziewcząt i mniej niż 1$ chłopców. Te dane mogą wskazywać na rosnącą świadomość zdrowotną wśród najmłodszych pokoleń oraz skuteczność programów edukacyjnych. Kontynuacja takich inicjatyw jest kluczowa dla dalszego ograniczenia skali zjawiska.  
5\. <b>Społeczno-ekonomiczne skutki palenia. </b> Uzależnienie od tytoniu generuje ogromne koszty społeczne i ekonomiczne. Obejmują one wydatki na opiekę zdrowotną, leczenie chorób wywołanych paleniem oraz utratę produktywności. Wprowadzenie bardziej rygorystycznych regulacji prawnych i dalsza promocja zdrowego stylu życia mogłaby zminimalizować te negatywne efekty.  
6\. <b>Potrzeba indywidualnie dostosowanych interwencji. </b> Dane sugerują, że polityki antynikotynowe powinny być bardziej ukierunkowane. Szczególną uwagę należy zwrócić na grupy, w których spadek odsetka palaczy jest mniej widoczny (np. mężczyźni w wieku 30-40 lat lub kobiety w grupach starszych). Skuteczne kampanie powinny uwzględnić specyfikę tych grup i ich motywacje do palenia.  

Podsumowując, rozprzestrzenianie się epidemii palenia papierosów w Polsce jest wciąż poważnym wyzwaniem zdrowotnym i społecznym, mimo zauważalnych pozytywnych zmian. Kluczowe dla dalszej poprawy sytuacji są zintegrowane działania obejmujące edukację, prewencję i wsparcie dla osób uzależnionych. Ważne jest, aby programy te były oparte na danych i dostosowane do specyficznych potrzeb różnych grup społecznych. Tylko dzięki takiemu podejściu możliwe będzie zredukowanie negatywnego wpływu palenia na społeczeństwo oraz gospodarkę, przyczyniając się jednocześnie do poprawy jakości życia obywateli.

## 7. Dodatki

### 7.1 Numeryczne sprawdzenie rzędu metody

W sekcji <i>4.3 Porównanie metod</i> stwierdziliśmy, że wykorzystamy metodę numeryczną Runge-Kutty w celu przybliżenia numerycznego wyników naszego modelu. Matematyczna analiza metody oparta na szeregu Taylora wskazuje, że rząd wynosi $p=4$. Będziemy jednak chcieli sprawdzić za pomocą narzędzi numerycznych, czy rzeczywiście rząd metody wyznaczony praktycznie zgadza się z matematycznym wyprowadzeniem. W tym celu wykorzystamy zmodyfikowaną <i> Definicję 1</i> związaną z globalnym błędem obcięcia dostosowując ją do praktycznego kontekstu poprzez wyliczenie maksymalnego błędu obcięcia $\left(\text{Max GTE}\right)$ oraz zależność wynikającą z <i>Faktu 1</i> dotyczącą rzędu metody. Wówczas definiujemy
\begin{align*}
    \text{Max GTE}_{h}=\max\left|y(t_n)-y_{n,h}\right| \\
    \text{Max GTE}_{\frac{h}{2}}=\max\left|y(t_n)-y_{n,\frac{h}{2}}\right|,
\end{align*}
gdzie $\text{Max GTE}_h$ oraz $\text{Max GTE}_\frac{h}{2}$ oznaczają maksymalny globalny błąd obcięcia dla kroków $h$ i $\frac{h}{2}$. Dodatkowo z <i> Faktu 1</i>
$$\text{GTE}\approx Ch^p,$$
gdzie $C$ jest stałą, a $h$ krokiem czasowym. W celu porównania błędów dla różnych kroków czasowych wykorzystamy zależność, że
$$\frac{\text{Max GTE}_h}{\text{Max GTE}_\frac{h}{2}}=\left(\frac{h}{\frac{h}{2}} \right)^p=2^p.$$
Stąd otrzymujemy rząd metody
$$p\approx \frac{\log\left(\frac{\text{Max GTE}_h}{\text{Max GTE}_\frac{h}{2}}\right)}{\log(2)}.$$
Poniższy kod wyznacza rząd metody dla każdego z równań naszego układu.

```python
import numpy as np

def dN_dt(t, N, O, M, B):
    return mu*P(N, O, M, B)-beta*N*((O+M)/P(N, O, M, B))-d_n*N

def dO_dt(t, N, O, M, B):
    return (beta*N*((O+M)/P(N, O, M, B))+alfa*M+rho*B*((O+M)/P(N, O, M, B))-gamma*O-lambd*M-d_o*O)

def dM_dt(t, N, O, M, B):
    return gamma*O-alfa*M-delta*M-d_m*M

def dB_dt(t, N, O, M, B):
    return lambd*O+delta*M-rho*B*((O+M)/P(N, O, M, B))-d_b*B

def P(N, O, M, B):
    return N+O+M+B

mu = 0.0101
beta = 0.443
d_n = 0.0108
d_o = 0.0216
d_m = 0.0324
d_b = 0.0140
alfa = 0.105
gamma = 0.243
lambd = 0.365
delta = 0.186
rho = 0.1215

def rk4(f, t, y, h):
    k1 = f(t, y) 
    k2 = f(t+0.5*h, y+0.5*h*k1)
    k3 = f(t+0.5*h, y+0.5*h*k2)
    k4 = f(t+h, y+h*k3)
    return y+(h/6)*(k1+2*k2+2*k3+k4)

def uklad(t, y):
    N, O, M, B = y
    dN = dN_dt(t, N, O, M, B)
    dO = dO_dt(t, N, O, M, B)
    dM = dM_dt(t, N, O, M, B)
    dB = dB_dt(t, N, O, M, B)
    return np.array([dN, dO, dM, dB])

def rozwiazanie_ukladu(t0, t_max, h, N0, O0, M0, B0):
    wartosc_t = []
    wartosc_y = []

    t = t0
    y = np.array([N0, O0, M0, B0])

    while t <= t_max:
        wartosc_t.append(t)
        wartosc_y.append(y)
        y = rk4(uklad, t, y, h)
        t += h

    return np.array(wartosc_t), np.array(wartosc_y)


def rzad_metody(t0, t_max, N0, O0, M0, B0):
    wartosc_h = [0.1, 0.05, 0.025] 
    blad = {"N": [], "O": [], "M": [], "B": []}
    
    for h in wartosc_h:
        t_h, y_h = rozwiazanie_ukladu(t0, t_max, h, N0, O0, M0, B0)
        t_h2, y_h2 = rozwiazanie_ukladu(t0, t_max, h/2, N0, O0, M0, B0)

        min_dl = min(len(y_h), len(y_h2[::2]))
        y_h = y_h[:min_dl]
        y_h_resampled = y_h2[::2][:min_dl]

        for i, var in enumerate(["N", "O", "M", "B"]):
            abs_blad = np.abs(y_h[:,i] - y_h_resampled[:,i])
            blad[var].append(np.max(abs_blad))

    for var in blad:
        wyniki = []
        for i in range(len(wartosc_h) - 1):
            wynik = np.log(blad[var][i] / blad[var][i+1]) / np.log(2)
            wyniki.append(wynik)
        avg_wynik = np.mean(wyniki)
        print(f"Rząd metody RK4 dla {var}: {avg_wynik:.2f}")

t0 = 0
t_max = 200
N0 = 25735469
O0 = 4106919
M0 = 4351982
B0 = 4225226

rzad_metody(t0, t_max, N0, O0, M0, B0)
```

### 7.2 Interaktywny wykres epidemii palenia papierosów w Polsce

Temat analizy zachowania populacji Polski w kontekście epidemii palenia papierosów jest bardzo interesujący, szczególnie w świetle zastosowania równań różniczkowych do modelowania dynamiki tego zjawiska. Model taki pozwala uwzględnić wiele zmiennych i parametrów, które mogą wpływać na rozwój epidemii. Poniżej znajduje się interaktywny wykres <i><b> (Wykres interaktywny) </b></i>, w którym można zmieniać wartości parametrów, co umożliwia przeprowadzanie dalszych analiz i lepsze zrozumienie dynamiki rozprzestrzeniania się nałogu w populacji.  
  
<i>Uwaga praktyczna: jeśli wykres interaktywny nie działał poprawnie należy skompilować go oddzielnie, nie za pomocą "Run All”. </i>

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

mu = 0.0101
d_n = 0.0108
d_o = 0.0216
d_m = 0.0324
d_b = 0.0140
beta = 0.443
alfa = 0.105
gamma = 0.243
lambd = 0.365
rho = 0.1215
delta = 0.186

def dN_dt(N, O, M, B, mu, beta, d_n):
    return mu*P(N, O, M, B)-beta*N*((O+M)/P(N, O, M, B))-d_n*N

def dO_dt(N, O, M, B, beta, alfa, rho, gamma, lambd, d_o):
    return beta*N*((O+M)/P(N, O, M, B))+alfa*M+rho*B*((O+M)/P(N, O, M, B))-gamma*O-lambd*M-d_o*O

def dM_dt(N, O, M, B, gamma, alfa, delta, d_m):
    return gamma*O-alfa*M-delta*M-d_m*M

def dB_dt(N, O, M, B, lambd, delta, rho, d_b):
    return lambd*O+delta*M-rho*B*((O+M)/P(N, O, M, B))-d_b*B

def P(N, O, M, B):
    return N+O+M+B

def rk4(f, t, y, h):
    k1 = f(t, y) 
    k2 = f(t+0.5*h, y+0.5*h*k1)
    k3 = f(t+0.5*h, y+0.5*h*k2)
    k4 = f(t+h, y+h*k3)
    return y+(h/6)*(k1+2*k2+2*k3+k4)

def rozwiazanie_ukladu(t0, t_max, h, N0, O0, M0, B0, mu, beta, alfa, gamma, lambd, rho, delta):
    wartosc_t = []
    wartosc_N = []
    wartosc_O = []
    wartosc_M = []
    wartosc_B = []

    t = t0
    N, O, M, B = N0, O0, M0, B0

    while t <= t_max:
        wartosc_t.append(t)
        wartosc_N.append(N)
        wartosc_O.append(O)
        wartosc_M.append(M)
        wartosc_B.append(B)

        N = rk4(lambda t, N: dN_dt(N, O, M, B, mu, beta, d_n), t, N, h)
        O = rk4(lambda t, O: dO_dt(N, O, M, B, beta, alfa, rho, gamma, lambd, d_o), t, O, h)
        M = rk4(lambda t, M: dM_dt(N, O, M, B, gamma, alfa, delta, d_m), t, M, h)
        B = rk4(lambda t, B: dB_dt(N, O, M, B, lambd, delta, rho, d_b), t, B, h)

        t += h

    return wartosc_t, wartosc_N, wartosc_O, wartosc_M, wartosc_B

def wykres_interaktywny(mu, beta, alfa, gamma, lambd, rho, delta):
    t0 = 0
    t_max = 200
    h = 0.1
    N0 = 25735469
    O0 = 4106919
    M0 = 4351982
    B0 = 4225226

    wartosc_t, wartosc_N, wartosc_O, wartosc_M, wartosc_B = rozwiazanie_ukladu(t0, t_max, h, N0, O0, M0, B0, mu, beta, alfa, gamma, lambd, rho, delta)

    plt.figure(figsize=(12, 8))

    plt.plot(wartosc_t, wartosc_N, label='N(t)', color='blue')
    plt.plot(wartosc_t, wartosc_O, label='O(t)', color='green')
    plt.plot(wartosc_t, wartosc_M, label='M(t)', color='red')
    plt.plot(wartosc_t, wartosc_B, label='B(t)', color='orange')

    plt.xlabel('Czas t')
    plt.ylabel('Wartości (skala logarytmiczna)')
    plt.yscale('log')
    plt.title('Wykres interkatywny')
    plt.legend()
    plt.grid()
    plt.show()

interact(wykres_interaktywny,
         mu=FloatSlider(min=0.001, max=0.02, step=0.0001, value=0.0101, readout_format='.3f'),
         beta=FloatSlider(min=0.1, max=1.0, step=0.001, value=0.443, readout_format='.3f'),
         alfa=FloatSlider(min=0.01, max=0.2, step=0.001, value=0.105, readout_format='.3f'),
         gamma=FloatSlider(min=0.1, max=0.5, step=0.001, value=0.243, readout_format='.3f'),
         lambd=FloatSlider(min=0.1, max=0.5, step=0.001, value=0.365, readout_format='.3f'),
         rho=FloatSlider(min=0.05, max=0.2, step=0.001, value=0.1215, readout_format='.3f'),
         delta=FloatSlider(min=0.1, max=0.5, step=0.001, value=0.186, readout_format='.3f'))
```
